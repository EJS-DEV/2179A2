```python
NZ = nz_tiles.avg_d_mbps.to_numpy()
AUS = au_tiles.avg_d_mbps.to_numpy()

NZ.sort()
AUS.sort()
fig = plt.figure(figsize=(5,5))
#ax = fig.add_subplot(2, 1, 1)
x1 = np.linspace(0,1,len(NZ))
x2 = np.linspace(0,1,len(AUS))


#ax.set_yscale('log')
plt.plot (x1, NZ, label="NZ")
plt.plot (x2, AUS, label="AUS")
plt.title("Comparing NZ/AUS Broadband Distribution - Q3 2023")
#plt.title("Speedtest Open Data")


plt.xlabel("Proportion of users")
plt.ylabel("Mbps")
plt.legend(loc="best")
plt.ylim(0,900)

plt.show()

# can plot multiple lines over same scale, with scaled x axis. ie BKK, MEL, LAX, SHG - PICKUP
# plot of log scale
```
